# 2.3 — Encoding Categorical Data

ML algorithms only understand numbers. Any text column must be converted into numbers before training.

> **Core question:** Does the category have an order?
> - No order (city, colour, gender) → **one-hot encoding**
> - Has order (small, medium, large) → **ordinal encoding**
> - Target column y → **label encoding**

### What is sklearn?
sklearn (scikit-learn) is the Python library that has all ML algorithms and data preparation tools ready to use. `LabelEncoder`, `OneHotEncoder`, `OrdinalEncoder` all come from sklearn.

```python
import pandas as pd
import numpy as np

data = {
    'name':   ['Arwindh', 'Priya', 'Ravi', 'Meena', 'Karan'],
    'city':   ['Chennai', 'Mumbai', 'Delhi', 'Mumbai', 'Chennai'],
    'size':   ['Small', 'Large', 'Medium', 'Small', 'Large'],
    'grade':  ['A', 'C', 'B', 'A', 'B'],
    'salary': [42000, 55000, 38000, 47000, 61000]
}

df = pd.DataFrame(data)
df
```

## 1. Label Encoding

Replaces each category with a number assigned **alphabetically**. Has no relationship with any other column.

- Chennai=0, Delhi=1, Mumbai=2 (just alphabetical order)
- Use only for the **target column y**, or tree-based models
- Never use on feature columns for linear models — creates false ordering

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df['grade_encoded'] = le.fit_transform(df['grade'])

print("Classes:", le.classes_)   # shows the mapping
df[['grade', 'grade_encoded']]
```

## 2. One-Hot Encoding

Creates a new binary column (0 or 1) for each unique category. No false ordering — every category is treated equally.

Use when: no natural order exists (city, colour, gender, department)

```python
# With pandas get_dummies — simplest way
df_ohe = pd.get_dummies(df, columns=['city'])
df_ohe
```

```python
# drop_first=True to avoid dummy variable trap
# if Chennai=0 and Delhi=0, model already knows it's Mumbai
df_ohe_drop = pd.get_dummies(df, columns=['city'], drop_first=True)
df_ohe_drop
```

```python
# With sklearn OneHotEncoder
from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder(sparse_output=False)
encoded = ohe.fit_transform(df[['city']])

print("Categories:", ohe.categories_)
print(encoded)
```

## 3. Ordinal Encoding

Use when categories have a meaningful order. You define the order explicitly — never let the encoder assign it alphabetically.

Use when: Small < Medium < Large, Low < Medium < High, etc.

```python
from sklearn.preprocessing import OrdinalEncoder

# Define the order explicitly
oe = OrdinalEncoder(categories=[['Small', 'Medium', 'Large']])
df['size_encoded'] = oe.fit_transform(df[['size']])

df[['size', 'size_encoded']]
```

```python
# Alternative — using pandas map (more readable)
order_map = {'Small': 0, 'Medium': 1, 'Large': 2}
df['size_mapped'] = df['size'].map(order_map)

df[['size', 'size_encoded', 'size_mapped']]
```

## 4. Putting It All Together

Encode all categorical columns in one go — the final DataFrame should have no text columns before passing to sklearn.

```python
df_final = df[['name', 'city', 'size', 'grade', 'salary']].copy()

# One-hot: city (no order)
df_final = pd.get_dummies(df_final, columns=['city'], drop_first=True)

# Ordinal: size (has order)
df_final['size'] = df_final['size'].map({'Small': 0, 'Medium': 1, 'Large': 2})

# Label: grade (target column)
df_final['grade'] = LabelEncoder().fit_transform(df_final['grade'])

# Drop name — not useful for ML
df_final = df_final.drop(columns=['name'])

df_final
```

```python
# Verify — check data types, no object columns should remain
df_final.dtypes
```

## 5. Decision Rules

| Situation | Encoding |
|-----------|----------|
| No natural order (city, colour, gender) | One-hot encoding |
| Has order (Small < Medium < Large) | Ordinal encoding |
| Target column y | Label encoding |
| Many unique values (50+ cities) | Target encoding (advanced — chapter 7) |
| Tree models (Random Forest, Decision Tree) | Label encoding is okay for features too |

> **Key insight:** Always check `df.dtypes` after encoding — if any column still shows `object`, it hasn't been encoded yet.
