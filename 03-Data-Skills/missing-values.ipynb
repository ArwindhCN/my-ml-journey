# 2.1 — Handling Missing Values

Missing values are cells in your dataset with no value. In pandas they show up as `NaN` (Not a Number).  
They happen because data wasn't collected, a survey was skipped, a sensor failed, etc.

> **Why it matters:** sklearn will throw an error if you pass a dataset with NaN values. You must handle them before training any model.

### Two choices:
- **Drop** — remove the row or column
- **Fill** — replace NaN with a sensible value

```python
import pandas as pd
import numpy as np

# Sample dataset with missing values
data = {
    'name':   ['Arwindh', 'Priya', 'Ravi', 'Meena', 'Karan'],
    'age':    [22, np.nan, 19, 23, np.nan],
    'salary': [45000, 62000, np.nan, 38000, np.nan],
    'city':   ['Chennai', 'Mumbai', np.nan, 'Delhi', 'Pune']
}

df = pd.DataFrame(data)
df
```

## 1. Detecting Missing Values

Always run this first on any new dataset.

```python
# Count of nulls per column
df.isnull().sum()
```

```python
# As a percentage
df.isnull().sum() / len(df) * 100
```

```python
# Which rows have at least one NaN?
df[df.isnull().any(axis=1)]
```

## 2. Dropping Missing Values

```python
# Drop any row that has at least one NaN
df.dropna()
```

```python
# Drop rows only if 'salary' is NaN (don't throw away rows for other columns)
df.dropna(subset=['salary'])
```

```python
# Drop a column that has too many missing values
df.drop(columns=['city'])
```

```python
# Keep rows that have at least 3 non-NaN values
df.dropna(thresh=3)
```

> **Note:** `dropna()` does not modify `df` — you must reassign: `df = df.dropna()`

## 3. Filling Missing Values

```python
# Fill numeric column with mean (good when data is normally distributed)
df['age'].fillna(df['age'].mean())
```

```python
# Fill numeric column with median (better when outliers exist)
df['salary'].fillna(df['salary'].median())
```

```python
# Fill categorical column with mode (most frequent value)
df['city'].fillna(df['city'].mode()[0])
```

```python
# Fill with a fixed constant
df['city'].fillna('Unknown')
```

```python
# Forward fill — copy value from previous row (useful for time series)
df['salary'].ffill()
```

## 4. Using sklearn's SimpleImputer (recommended for ML pipelines)

```python
from sklearn.impute import SimpleImputer

# For numeric columns
imp_num = SimpleImputer(strategy='median')
df[['age', 'salary']] = imp_num.fit_transform(df[['age', 'salary']])

# For categorical columns
imp_cat = SimpleImputer(strategy='most_frequent')
df[['city']] = imp_cat.fit_transform(df[['city']])

df
```

```python
# Verify — no NaNs should remain
df.isnull().sum()
```

## 5. Decision Rules

| Situation | Strategy |
|-----------|----------|
| Missing < 5% of rows | Drop the rows |
| Column missing > 50-70% | Drop the column |
| Numeric, no outliers | Fill with mean |
| Numeric, outliers exist | Fill with median |
| Categorical column | Fill with mode |
| "Missing" is meaningful | Fill with constant (e.g. 0, 'Unknown') |
| Time series data | Forward fill / backward fill |

> **Key insight:** Always think about *why* the value is missing. That usually tells you the best way to handle it.
