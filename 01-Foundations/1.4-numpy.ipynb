> Core idea: NumPy is Python's fast number-crunching engine. Every ML library (pandas, sklearn, matplotlib) is built on top of it.

## Why NumPy?

Python lists are slow for maths. If you have 1 million data points and want to multiply each by 2:

```python
# Python list — slow loop internally
result = [x * 2 for x in my_list]

# NumPy — C-speed, one operation
result = np.array(my_list) * 2
```

NumPy is **10-100x faster** than pure Python for numerical operations. That's why it's the foundation of all ML.

## Analogy — A Calculator vs a Spreadsheet

Python list = a basic calculator — one number at a time.
NumPy array = a spreadsheet — apply a formula to an entire column in one click.

````python
import numpy as np

# Creating arrays
a = np.array([1, 2, 3, 4, 5])          # from list
b = np.zeros(5)                          # all zeros
c = np.ones((3, 4))                      # 3x4 matrix of ones
d = np.arange(0, 10, 2)                  # 0,2,4,6,8
e = np.linspace(0, 1, 5)                 # 5 evenly spaced between 0 and 1
f = np.random.randn(3, 3)               # 3x3 random normal

print("Array from list:", a)
print("Zeros:", b)
print("arange:", d)
print("linspace:", e)
print("\nRandom 3x3:\n", f)
````

````python
# Array properties — always check these on new data
X = np.array([
    [25000, 30, 1],
    [45000, 25, 0],
    [35000, 40, 1],
    [60000, 28, 0]
])

print("Shape:", X.shape)     # (rows, cols)
print("Dimensions:", X.ndim) # 2D
print("Data type:", X.dtype) # int64 or float64
print("Size:", X.size)       # total elements
````

````python
# Indexing and slicing
print("First row:", X[0])             # row 0
print("Last row:", X[-1])             # last row
print("First column:", X[:, 0])       # all rows, col 0
print("Element [1,2]:", X[1, 2])      # row 1, col 2
print("First 2 rows:\n", X[:2, :])    # rows 0 and 1
print("Last 2 cols:\n", X[:, -2:])    # last 2 columns
````

````python
# Arithmetic — all element-wise
incomes = X[:, 0].astype(float)   # first column = incomes

print("Incomes:", incomes)
print("After tax (30%):", incomes * 0.7)
print("In lakhs:", incomes / 100000)
print("Sqrt:", np.sqrt(incomes))
````

````python
# Statistical operations on arrays
print("Mean income: ₹", np.mean(incomes))
print("Median:      ₹", np.median(incomes))
print("Std Dev:     ₹", np.std(incomes))
print("Min / Max:    ₹", incomes.min(), "/", incomes.max())

# Column-wise (per feature)
print("\nColumn means:", X.mean(axis=0))
print("Row sums:", X.sum(axis=1))
````

````python
# Boolean masking — filter rows
incomes_all = np.array([25000, 45000, 35000, 60000, 22000, 80000])

mask = incomes_all > 35000
print("Mask:", mask)
print("High earners:", incomes_all[mask])

# How many?
print("Count above 35k:", np.sum(mask))
````

````python
# Reshaping — very common in ML pipelines
flat = np.array([1, 2, 3, 4, 5, 6])
print("Original shape:", flat.shape)

matrix = flat.reshape(2, 3)       # 2 rows, 3 cols
print("Reshaped to (2,3):\n", matrix)

# -1 means 'figure it out'
col = flat.reshape(-1, 1)         # column vector
print("Column vector shape:", col.shape)  # (6, 1)
````

````python
# Broadcasting — NumPy's superpower
# Subtract column mean from each column (manual standardisation)

data = np.array([
    [170, 65],   # height cm, weight kg
    [160, 55],
    [180, 80],
    [175, 70]
], dtype=float)

col_means = data.mean(axis=0)     # mean of each column
col_stds  = data.std(axis=0)      # std of each column

standardised = (data - col_means) / col_stds  # broadcasting — applies to all rows
print("Column means:", col_means)
print("\nStandardised data (what StandardScaler does):\n", standardised.round(3))
````

## Most used NumPy in ML — Quick Reference

```python
np.array([...])          # create array
arr.shape                # dimensions
arr.reshape(-1, 1)       # make column vector (needed by sklearn)
arr[:, 0]                # select column
arr[arr > 0]             # filter by condition
np.mean / median / std   # stats
np.dot(a, b)             # dot product
a @ b                    # matrix multiply
np.log1p(arr)            # log transform
np.where(cond, a, b)     # conditional replacement
np.concatenate([a, b])   # join arrays
np.random.seed(42)       # reproducibility
```
