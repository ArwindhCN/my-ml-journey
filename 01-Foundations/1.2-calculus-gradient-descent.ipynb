# 1.2 — Calculus
**Derivatives & Gradient Descent**

> Core idea: The model starts with random weights and needs to improve. Calculus tells it which direction to adjust and by how much.

## Why ML needs Calculus

From 1.1 we know: prediction = X · W

But how does the model find the **right weights W**? It doesn't know them. It starts with random numbers and uses calculus to **walk toward better weights** — step by step — until predictions are good.

Without calculus, training a model would be guesswork.

## Analogy — Blindfolded on a Hill

Imagine you're **blindfolded in hilly terrain** and your goal is the lowest valley.

- You can't see anything
- But you can **feel the slope** under your feet
- If it slopes down to the left → step left
- If it slopes down to the right → step right  
- If it's flat → you're at the bottom. Stop.

The **hill** = model's error. The **slope** = derivative. The **step** = weight update. The **valley** = minimum error.

## Key Concepts

### Derivative
The slope of a curve at any point — how fast something is changing.

```
f(x)  = x²
f'(x) = 2x    ← derivative
```

At x=3 → slope = 6 (steep, going up fast)  
At x=0 → slope = 0 (flat — this is the **minimum**)

**Key insight:** Where derivative = 0, you're at the minimum. ML wants minimum error.

### Loss Function
Measures how wrong the model is. The most common one:

```
Loss = (1/n) × Σ(actual - predicted)²   ← Mean Squared Error
```

Higher loss = further up the hill. Goal = bring loss to its minimum.

### Gradient Descent Update Rule

```
new_weight = old_weight - (learning_rate × gradient)
```

| Part | Meaning |
|---|---|
| `old_weight` | Current position on the hill |
| `gradient` | Direction of uphill (derivative of loss) |
| `learning_rate` | Step size |
| subtract | We go OPPOSITE to uphill = downhill |

### Learning Rate
- Too large → overshoot, bounce around, never settle
- Too small → takes forever
- Just right → smooth descent to minimum

Like tuning an old radio: too hard = skip past the station, too gentle = never find it.

````python
import numpy as np
import matplotlib.pyplot as plt

# Visualise the loss surface — a simple parabola (x²)
x = np.linspace(-3, 3, 100)
loss = x**2

plt.figure(figsize=(8, 4))
plt.plot(x, loss, color='steelblue', linewidth=2)
plt.axvline(x=0, color='green', linestyle='--', label='Minimum (slope=0)')
plt.scatter([2], [4], color='red', s=100, zorder=5, label='Starting point')
plt.title("The Loss Surface — we want the valley")
plt.xlabel("Weight value")
plt.ylabel("Loss (error)")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
````

````python
# Gradient Descent from scratch
# Dataset: house sizes (sqft) vs prices (₹ lakhs)
# True relationship: price = 2 × size

X = np.array([10, 20, 30, 40, 50], dtype=float)  # size in hundreds sqft
y = np.array([20, 40, 60, 80, 100], dtype=float) # price in lakhs

# Start with random weight
w = 0.0
learning_rate = 0.001
epochs = 200

loss_history = []
weight_history = []

for epoch in range(epochs):
    predictions = w * X                              # Step 1: predict
    loss = np.mean((y - predictions) ** 2)           # Step 2: MSE loss
    gradient = -2 * np.mean((y - predictions) * X)  # Step 3: gradient
    w = w - learning_rate * gradient                 # Step 4: update weight
    
    loss_history.append(loss)
    weight_history.append(w)

print(f"Final weight: {w:.4f}  (true value = 2.0)")
print(f"Final loss:   {loss_history[-1]:.6f}")
````

````python
# Plot the loss dropping over epochs
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))

ax1.plot(loss_history, color='coral')
ax1.set_title("Loss dropping over epochs")
ax1.set_xlabel("Epoch")
ax1.set_ylabel("Loss (MSE)")
ax1.grid(True, alpha=0.3)

ax2.plot(weight_history, color='steelblue')
ax2.axhline(y=2.0, color='green', linestyle='--', label='True weight = 2.0')
ax2.set_title("Weight converging to 2.0")
ax2.set_xlabel("Epoch")
ax2.set_ylabel("Weight value")
ax2.legend()
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
````

````python
# Effect of different learning rates
learning_rates = [0.0001, 0.001, 0.01, 0.1]
colors = ['blue', 'green', 'orange', 'red']
labels = ['Too slow (0.0001)', 'Just right (0.001)', 'Aggressive (0.01)', 'Too large (0.1)']

plt.figure(figsize=(10, 5))

for lr, color, label in zip(learning_rates, colors, labels):
    w_temp = 0.0
    losses = []
    for _ in range(200):
        preds = w_temp * X
        loss = np.mean((y - preds) ** 2)
        grad = -2 * np.mean((y - preds) * X)
        w_temp = w_temp - lr * grad
        losses.append(min(loss, 5000))  # cap for visibility
    plt.plot(losses, color=color, label=label)

plt.title("Effect of Learning Rate on Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()
plt.ylim(0, 3000)
plt.grid(True, alpha=0.3)
plt.show()
````

## What happens during training — step by step

```
1. Start with random weights W
2. Predict: ŷ = X · W
3. Calculate loss: how wrong are we?
4. Calculate gradient: which way is uphill?
5. Update: W = W - learning_rate × gradient
6. Repeat 2-5 for many epochs
7. Stop when loss stops decreasing
```

Each full pass through the data = one **epoch**.

## Where this appears in real ML

| Algorithm | Uses gradient descent? |
|---|---|
| Logistic Regression | Yes |
| Neural Networks | Yes (core of backprop) |
| Linear Regression | Yes (or closed-form) |
| Decision Trees | No (uses splits instead) |
| KNN | No (no training phase) |
