# 1.1 — Linear Algebra
**Vectors, Matrices, Dot Products**

> Core idea: All data in ML is numbers in grids. Linear algebra is how we work with those grids efficiently — without slow loops.

## Why ML needs Linear Algebra

Every piece of data you feed a machine — a photo, a loan application, a patient record — gets converted into **rows and columns of numbers**.

Linear algebra is the mathematics of working with those grids. Without it:
- You'd loop through every row one by one (very slow)
- You couldn't represent a model's learned knowledge compactly
- You couldn't do predictions for 10,000 rows at once

With it: one operation, all rows, instant.

## Analogy — The Train Reservation Chart

Imagine a **train reservation chart**:
- Each **row** = one passenger
- Each **column** = a detail: age, ticket class, fare, distance

That entire chart is a **matrix**.
One passenger's row alone is a **vector**.

When the model processes all passengers at once to find patterns → that's **matrix operations**.

## Key Concepts

| Concept | Plain English | ML use |
|---|---|---|
| **Vector** | A list of numbers = one data point | One row of your dataset |
| **Matrix** | A grid of numbers = your full dataset | All rows and columns |
| **Dot product** | Multiply element-wise, then sum | Core of every prediction |
| **Matrix multiply** | Apply model to all rows at once | predictions = X · W |

## The Core Insight

Every ML prediction is a **dot product**:

```
prediction = (age × w1) + (fare × w2) + (class × w3) + ...
           = data_vector · weight_vector
```

The model learns the weights. You supply the data. The dot product gives the prediction.

````python
import numpy as np

# --- VECTORS ---
# One Titanic passenger: [age, survived, class, fare]
passenger = np.array([24, 1, 3, 7.25])
print("Passenger vector:", passenger)
print("Shape:", passenger.shape)  # (4,) = 1D with 4 elements
````

````python
# --- MATRICES ---
# Multiple passengers = matrix (rows=passengers, cols=features)
X = np.array([
    [24, 1, 3,  7.25],
    [38, 0, 1, 71.30],
    [26, 1, 2, 12.00],
    [35, 0, 3,  8.05]
])
print("Data matrix shape:", X.shape)  # (4 passengers, 4 features)
print("\nMatrix:\n", X)
````

````python
# --- DOT PRODUCT ---
# Model weights (what the model has learned)
weights = np.array([0.1, 0.5, 0.3, 0.02])

# Prediction for ONE passenger = dot product
prediction = np.dot(passenger, weights)
print("Single prediction (dot product):", prediction)

# Manual version (same result):
manual = (24*0.1) + (1*0.5) + (3*0.3) + (7.25*0.02)
print("Manual version:", manual)
````

````python
# --- MATRIX MULTIPLICATION ---
# Predictions for ALL passengers at once
all_predictions = X @ weights  # @ = matrix multiply in numpy
print("Predictions for all passengers:", all_predictions)
# One number per passenger — that's the model's output for each row
````

````python
# --- USEFUL MATRIX OPERATIONS ---

# Transpose: flip rows and columns
print("Original shape:", X.shape)       # (4, 4)
print("Transposed shape:", X.T.shape)   # (4, 4) — square here, but concept is clear

# Element-wise operations
X_doubled = X * 2
print("\nDoubled matrix:\n", X_doubled)

# Sum along axis
print("\nColumn means (mean of each feature):", X.mean(axis=0))
print("Row sums (sum across features per passenger):", X.sum(axis=1))
````

## When does this appear in real ML?

- **Every time you call `model.predict(X)`** — internally it's doing X · W
- **Neural networks** — stack of matrix multiplications
- **Feature scaling** — matrix operations on entire columns
- **PCA** (Chapter 6) — built entirely on matrix decomposition

## Summary

| Thing | What it is | Why it matters |
|---|---|---|
| Vector | 1D array — one row | Represents one data point |
| Matrix | 2D array — all rows | Your entire dataset X |
| Dot product | Element multiply + sum | Core of every prediction |
| `X @ W` | Matrix × vector | Predict all rows at once |
